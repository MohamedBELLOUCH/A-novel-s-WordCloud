```python
def read_pages(n, m):
  assert n <= m and n > 1
  s = ""
  pdfFileObj = open('/content/hemingway.pdf', 'rb')
  pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
  for p in range(n, m+1):
    pageObj = pdfReader.getPage(p)
    s += pageObj.extractText()
  pdfFileObj.close()
  return s
```

```python
txt = read_pages(20, 60)
```

```python
len(txt)
```

```python
from wordcloud import WordCloud, STOPWORDS
stopwords = set(STOPWORDS)
stopwords = stopwords.union({'now', 'thought', 'will'})
```

```python
hemingway = WordCloud(
    background_color='white',
    stopwords=stopwords
)
hemingway.generate(txt)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.imshow(hemingway, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
def img_on_img(img1, img2):
  img = img1.copy()
  for i in range(img.shape[0]):
    for j in range(img.shape[1]):
      if img2[i, j][0] != 255 or img2[i, j][1] != 255 or img2[i, j][1] != 255:
        img[i, j] = img2[i, j]
  return img
```
